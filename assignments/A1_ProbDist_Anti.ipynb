```python
import matplotlib.pyplot as plt
# Within IPython notebook, add multiple print capabilities
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
# Print below code
%matplotlib inline
```

```python
import numpy as np
```

<h2>Discrete: Uniform Distribution</h2>

```python
# x (events) and y (probabilities)
evts = np.arange(1,7)
prob = np.array([1/6]*6)
print("Events","\tProbabilities")
for i,j in zip(evts,prob):
    print(i,"\t",round(100*j,ndigits=1),"%")

# scatterplot
fig, ax = plt.subplots(1,1,figsize=(6,4))
ax.scatter(evts,prob,color='darkblue',linewidths=5)

# set display
ax.set_title('UNIFORM')
ax.set_xbound(0,7); ax.set_ybound(0,0.5)
ax.grid(color='lightgrey', linestyle='--', linewidth=1)
```

<h2>Discrete: Bernoulli Distribution</h2>

```python
# x (events) and y (probabilities)
poll = {'Biden':.522,'Trump':.419}
base = poll['Biden'] + poll ['Trump']
evts = ['Biden','Trump']
prob = [(poll['Biden']/base),(poll['Trump']/base)]
print("Events","\tProbabilities")
for i,j in zip(evts,prob):
    print(i,"\t",round(100*j,ndigits=1),"%")

# barplot
fig, ax = plt.subplots(1,1,figsize=(6,4))
ax.bar(evts,prob,color='blue',width=0.6)[1].set_color('red')

# set display
ax.set_title('BERNOULLI')
ax.set_ybound(0,1)
ax.grid(color='lightgrey', linestyle='--', linewidth=1)
```

<h2>Continuous: Logistic Distribution</h2>

```python
from scipy.stats import logistic
```

```python
# array for function argument 
x = np.linspace(-3,3,99)
```

<h3>Logistic Distribution PDF</h3>

```python
# series with various parameters
x1 = logistic.pdf(x,loc=0,scale=1)
x2 = logistic.pdf(x,loc=0,scale=0.7)
x3 = logistic.pdf(x,loc=-1,scale=1)
x4 = logistic.pdf(x,loc=1,scale=0.6)
x5 = logistic.pdf(x,loc=1,scale=1.2)

# plot
fig, ax = plt.subplots(1,1,figsize=(10,6))
ax.plot(x,x1,color='black',ls='-',lw=3,label='m=0, s=1.0')
ax.plot(x,x2,color='green',ls='-',label='m=0, s=0.7')
ax.plot(x,x3,color='orange',ls='--',label='m=-1, s=1.0')
ax.plot(x,x4,color='darkblue',ls=':',label='m=1, s=0.6')
ax.plot(x,x5,color='red',ls=':',label='m=1, s=1.2')
plt.legend(loc=2)
```

<h3>Logistic Distribution CDF</h3>

```python
# series with various parameters
x1c = logistic.cdf(x,loc=0,scale=1)
x2c = logistic.cdf(x,loc=0,scale=0.7)
x3c = logistic.cdf(x,loc=-1,scale=1)
x4c = logistic.cdf(x,loc=1,scale=0.6)
x5c = logistic.cdf(x,loc=1,scale=1.2)

# plot
fig, ax = plt.subplots(1,1,figsize=(10,6))
ax.plot(x,x1c,color='black',ls='-',lw=3,label='m=0, s=1.0')
ax.plot(x,x2c,color='green',ls='-',label='m=0, s=0.7')
ax.plot(x,x3c,color='orange',ls='--',label='m=-1, s=1.0')
ax.plot(x,x4c,color='darkblue',ls=':',label='m=1, s=0.6')
ax.plot(x,x5c,color='red',ls=':',label='m=1, s=1.2')
plt.legend(loc=2)
```
